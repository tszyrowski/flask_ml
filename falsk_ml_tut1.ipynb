```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
import pandas as pd
import io
import os
import requests
import numpy as np
from sklearn import metrics
```

```python
df  = pd.read_csv(r"auto-mpg.csv", na_values=["NA", "?"])
df.head()
```

```python
df.isnull().sum()
```

```python
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())
df["mpg"] = df["mpg"].fillna(df["mpg"].median())
```

```python
df.isnull().sum()
```

```python
x = df[["cylinders", "displacement", "horsepower",	"weight", "acceleration", "model year",	"origin"]].values
y = df["mpg"].values
```

```python
y
```

```python
seed_state = 100
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=seed_state)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model = Sequential()
model.add(Dense(25, input_dim=x.shape[1], activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam")

monitor = EarlyStopping(monitor="val_loss", min_delta=1e-3, patience=5, verbose=1, mode="auto", restore_best_weights=True)

model.fit(x_train, y_train, validation_data=(x_test, y_test), callbacks=[monitor], verbose=2, epochs=1000)
```

```python
prediction = model.predict(x_test)
score = np.sqrt(metrics.mean_squared_error(prediction, y_test))
print(f"With seed: {seed_state} The score is: {score}")
```

```python
model.save(os.path.join(os.getcwd(), f"mog_model_{seed_state}_{score}.h5"))
```
